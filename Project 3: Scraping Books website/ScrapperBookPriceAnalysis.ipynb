```python
import requests
from bs4 import BeautifulSoup

# URL of the BooksWagon website
url = "https://www.bookswagon.com/"

# Send an HTTP request to the website
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract specific information based on HTML structure
    # Example: Extracting book titles
    book_titles = soup.find_all('h2', class_='title')
    for title in book_titles:
        print(title.text)

    # Example: Extracting book prices
    book_prices = soup.find_all('div', class_='price')
    for price in book_prices:
        print(price.text)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
